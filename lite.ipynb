```python
import os
import random
import shutil

# Ruta de la carpeta datasets/
base_path = 'datasets/'

# Ruta de la carpeta de destino relativa al directorio actual
destination_base = 'datasetsLITE/'

def copy_folder_structure(src, dst):
    os.makedirs(dst, exist_ok=True)
    for item in os.listdir(src):
        item_path = os.path.join(src, item)
        if os.path.isdir(item_path):
            copy_folder_structure(item_path, os.path.join(dst, item))

# Obtener la lista de carpetas en la carpeta base
carpetas_base = os.listdir(base_path)

# Recorrer cada carpeta
for carpeta in carpetas_base:
    # Ruta de la carpeta de imágenes
    images_folder = os.path.join(base_path, carpeta, 'images', 'train')
    # Ruta de la carpeta de etiquetas
    labels_folder = os.path.join(base_path, carpeta, 'labels', 'train')
    
    # Verificar si las carpetas de imágenes y etiquetas existen
    if os.path.exists(images_folder) and os.path.exists(labels_folder):
        # Crear la carpeta de destino para la carpeta actual
        destination_folder = os.path.join(destination_base, carpeta)
        copy_folder_structure(images_folder, os.path.join(destination_folder, 'images'))
        copy_folder_structure(labels_folder, os.path.join(destination_folder, 'labels'))
        
        # Obtener la lista de archivos jpg en la carpeta de imágenes
        jpg_files = [file for file in os.listdir(images_folder) if file.endswith('.jpg')]
        
        # Seleccionar solo 10 archivos al azar
        selected_files = random.sample(jpg_files, k=min(10, len(jpg_files)))
        
        # Copiar los archivos seleccionados a la carpeta de destino
        for file in selected_files:
            image_src = os.path.join(images_folder, file)
            image_dst = os.path.join(destination_folder, 'images', file)
            shutil.copy(image_src, image_dst)
            
            label_src = os.path.join(labels_folder, file[:-4] + '.txt')
            label_dst = os.path.join(destination_folder, 'labels', file[:-4] + '.txt')
            
            if os.path.exists(label_src):
                shutil.copy(label_src, label_dst)
```

```python
import os
import random
import shutil

def copy_dataset_structure(base_path, destination_base, num_train_files, num_test_files, num_valid_files):
    def copy_folder_structure(src, dst):
        os.makedirs(dst, exist_ok=True)
        for item in os.listdir(src):
            item_path = os.path.join(src, item)
            if os.path.isdir(item_path):
                copy_folder_structure(item_path, os.path.join(dst, item))

    carpetas_base = [carpeta for carpeta in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, carpeta))]

    for carpeta in carpetas_base:
        images_folder = os.path.join(base_path, carpeta, 'images')
        labels_folder = os.path.join(base_path, carpeta, 'labels')

        if os.path.exists(images_folder) and os.path.exists(labels_folder):
            destination_folder = os.path.join(destination_base, carpeta)
            os.makedirs(destination_folder, exist_ok=True)
            os.makedirs(os.path.join(destination_folder, 'images', 'train'), exist_ok=True)
            os.makedirs(os.path.join(destination_folder, 'images', 'test'), exist_ok=True)
            os.makedirs(os.path.join(destination_folder, 'images', 'valid'), exist_ok=True)
            os.makedirs(os.path.join(destination_folder, 'labels', 'train'), exist_ok=True)
            os.makedirs(os.path.join(destination_folder, 'labels', 'test'), exist_ok=True)
            os.makedirs(os.path.join(destination_folder, 'labels', 'valid'), exist_ok=True)

            train_folder = os.path.join(images_folder, 'train')
            test_folder = os.path.join(images_folder, 'test')
            valid_folder = os.path.join(images_folder, 'valid')

            train_files = [file for file in os.listdir(train_folder) if file.endswith('.jpg')]
            test_files = [file for file in os.listdir(test_folder) if file.endswith('.jpg')]
            valid_files = [file for file in os.listdir(valid_folder) if file.endswith('.jpg')]

            train_selected = random.sample(train_files, k=min(num_train_files, len(train_files)))
            test_selected = random.sample(test_files, k=min(num_test_files, len(test_files)))
            valid_selected = random.sample(valid_files, k=min(num_valid_files, len(valid_files)))

            for file in train_selected:
                image_src = os.path.join(train_folder, file)
                image_dst = os.path.join(destination_folder, 'images', 'train', file)
                shutil.copy(image_src, image_dst)

                label_src = os.path.join(labels_folder, 'train', file[:-4] + '.txt')
                label_dst = os.path.join(destination_folder, 'labels', 'train', file[:-4] + '.txt')
                if os.path.exists(label_src):
                    shutil.copy(label_src, label_dst)

            for file in test_selected:
                image_src = os.path.join(test_folder, file)
                image_dst = os.path.join(destination_folder, 'images', 'test', file)
                shutil.copy(image_src, image_dst)

                label_src = os.path.join(labels_folder, 'test', file[:-4] + '.txt')
                label_dst = os.path.join(destination_folder, 'labels', 'test', file[:-4] + '.txt')
                if os.path.exists(label_src):
                    shutil.copy(label_src, label_dst)

            for file in valid_selected:
                image_src = os.path.join(valid_folder, file)
                image_dst = os.path.join(destination_folder, 'images', 'valid', file)
                shutil.copy(image_src, image_dst)

                label_src = os.path.join(labels_folder, 'valid', file[:-4] + '.txt')
                label_dst = os.path.join(destination_folder, 'labels', 'valid', file[:-4] + '.txt')
                if os.path.exists(label_src):
                    shutil.copy(label_src, label_dst)

    copy_folder_structure(base_path, destination_base)
```
