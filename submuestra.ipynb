```python
import os
import shutil
import random

# Directorios de origen
src_dirs = {
    "images": {
        "train": "NewDataset/MOTORBIKE/images/train",
        "valid": "NewDataset/MOTORBIKE/images/valid",
        "test": "NewDataset/MOTORBIKE/images/test"
    },
    "labels": {
        "train": "NewDataset/MOTORBIKE/labels/train",
        "valid": "NewDataset/MOTORBIKE/labels/valid",
        "test": "NewDataset/MOTORBIKE/labels/test"
    }
}

# Directorios de destino (aquí deberías especificar tus propios directorios de destino)
dst_dirs = {
    "images": {
        "train": "NewDataset/Muestra/images/train",
        "valid": "NewDataset/Muestra/images/valid",
        "test": "NewDataset/Muestra/images/test"
    },
    "labels": {
        "train": "NewDataset/Muestra/labels/train",
        "valid": "NewDataset/Muestra/labels/valid",
        "test": "NewDataset/Muestra/labels/test"
    }
}

# Crear directorios de destino si no existen
for key in dst_dirs:
    for dataset in ["train", "valid", "test"]:
        os.makedirs(dst_dirs[key][dataset], exist_ok=True)

# Número de anotaciones para cada conjunto
num_annotations = {
    "train": 30,
    "valid": 6,
    "test": 12,
}

# Iterar sobre cada conjunto (train, valid, test)
for dataset in ["train", "valid", "test"]:
    count = 0  # Contador de anotaciones
    while count < num_annotations[dataset]:
        # Seleccionar un archivo .txt aleatoriamente
        txt_file = random.choice(os.listdir(src_dirs["labels"][dataset]))

        # Leer el archivo y obtener las líneas (anotaciones)
        with open(os.path.join(src_dirs["labels"][dataset], txt_file), "r") as f:
            lines = f.readlines()

        # Si el archivo tiene más anotaciones de las necesarias, saltarlo
        if len(lines) + count > num_annotations[dataset]:
            continue

        # Incrementar el contador
        count += len(lines)

        # Copiar el archivo .txt al directorio de destino
        shutil.copy(os.path.join(src_dirs["labels"][dataset], txt_file),
                    os.path.join(dst_dirs["labels"][dataset], txt_file))

        # Copiar el archivo de imagen correspondiente al directorio de destino
        img_file = txt_file.replace(".txt", ".jpg")
        shutil.copy(os.path.join(src_dirs["images"][dataset], img_file),
                    os.path.join(dst_dirs["images"][dataset], img_file))
```
